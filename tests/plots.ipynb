```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("/home/AD/yul080/ckpts/log.txt","r") as f:
    data_8 = f.readlines()

with open("/home/AD/yul080/DMLP_test_with_ddpm/log.txt","r") as f:
    data_128 = f.readlines()
```

```python
data_8 = data_8[6:]
data_128 = data_128[6:]
```

```python
def f(datum):
    if not ('Saving' in datum or 'saving' in datum):
        datum = datum.replace(";","").split(" ")
        return [int(datum[1]),float(datum[5]),float(datum[11])]
```

```python
data_8 = np.array([f(i) for i in data_8 if f(i) is not None and f(i)[0]<=96099])
data_128 = np.array([f(i) for i in data_128 if f(i) is not None and f(i)[0]<=96099])
```

```python
max(data_128[:,0])
```

```python
plt.plot(data_8[:,0],data_8[:,1], label = "batch size 8")
plt.plot(data_128[:,0],data_128[:,1], label = "batch size 128")
plt.xlabel("iteration")
plt.ylabel("training loss")
plt.title("training loss by iteration")
plt.yscale("log")
plt.legend()
```

```python
plt.plot(data_8[:,0],data_8[:,2], label = "batch size 8")
plt.plot(data_128[:,0],data_128[:,2], label = "batch size 128")
plt.xlabel("iteration")
plt.ylabel("ddpm loss")
plt.title("ddpm loss by iteration")
plt.yscale("log")
plt.legend()
```
